```python
import subprocess

def get_connected_gpus():
    try:
        result = subprocess.run(["nvidia-smi", "--list-gpus"], capture_output=True, text=True)
        if result.returncode == 0:
            gpu_list = result.stdout.strip().split('\n')
            return gpu_list
        else:
            return None
    except FileNotFoundError:
        print("nvidia-smi not found. Make sure NVIDIA GPU drivers are installed.")
        return None

if __name__ == "__main__":
    connected_gpus = get_connected_gpus()
    if connected_gpus is not None:
        for i, gpu in enumerate(connected_gpus):
            print(f"GPU {i}: {gpu}")
    else:
        print("Unable to retrieve GPU information.")
```

```python
from ultralytics import YOLO
import torch

# Initialize the YOLO model
model = YOLO()
model = YOLO("yolov8n.pt")

# Train the model using the provided data and for the specified number of epochs, using the GPU on your Mac
if torch.cuda.is_available():
  device = "cuda"
else:
  device = "cpu"

model.train(data="/kaggle/input/barcode-detection/data.yaml", optimizer='AdamW',
            epochs=20, device=device)
```
